````python
# 2 ë°°ìˆ˜ = 5ë¥¼ ë”í•´ê°€ëŠ” ê²ƒ(0ë¶€í„°)
num = 0
answer = []

while True:
    if num >= 100:
        break
    num += 5
    answer.append(num)
    
print(answer)
````

````python
# 3 ë°°ìˆ˜ = 5ë¥¼ ë”í•´ê°€ëŠ” ê²ƒ(0ë¶€í„°)
num = 5
answer = []

while True:
    answer.append(num)
    if num >= 100:
        break
    num += 5
    
print(answer)
````

### 13ë²ˆ. ë”•ì…”ë„ˆë¦¬ì—ì„œ ê°€ì¥ ë¹„ì‹¼ ê³¼ì¼ ì°¾ê¸°


````python
fruits = {'apple': 1000, 'banana': 500, 'orange': 800}
max_price = 0
max_fruit = ''

for fruit, price in fruits.items():
    if max_price < price:
        max_price = price
        max_fruit = fruit

print(max_fruit, max_price)
````

````python
fruits = {'apple': 1000, 'banana': 500, 'orange': 800}
expensive = ''

for key in fruits:
    if fruits.get(expensive, 0) < fruits[key]:
        expensive = key

print(expensive, fruits[expensive])
````

````python
fruits = {'apple': 1000, 'banana': 500, 'orange': 800, 'lemon': 1000}
max_price = 0
max_fruit = []

for fruit, price in fruits.items():
    
    # ìµœê³ ê°€ê°€ ê²½ì‹ ë˜ì—ˆì„ ë•Œ:
    if max_price < price:
        max_price = price
        max_fruit = [fruit]
    # ìµœê³ ê°€ê°€ ë˜ ë‚˜ì™”ì„ ë•Œ:
    elif max_price == price:
        max_fruit.append(fruit)
        
print(max_fruit, max_price)
````

````python
fruits = {'apple': 1000, 'banana': 500, 'orange': 800, 'lemon': 1000}
fruit_price = []
expensive = []

for fruit in fruits:
    fruit_price.append([fruits[fruit], fruit])

fruit_price.sort()

for i in range(len(fruit_price)):
    if fruit_price[-1][0] == fruit_price[i][0]:
        expensive.append(fruit_price[i][1])

print(expensive, fruits[expensive[0]])
````

### 15ë²ˆ. ë¬¸ìì—´ì—ì„œ ê° ë¬¸ì ê°œìˆ˜ ì„¸ê¸°

ì‚¬ìš©ìë¡œë¶€í„° ë¬¸ìì—´ì„ ì…ë ¥ë°›ì•„, ê° ë¬¸ìê°€ ëª‡ ë²ˆ ë‚˜ì˜¤ëŠ”ì§€ ë”•ì…”ë„ˆë¦¬ í˜•íƒœë¡œ ì¶œë ¥í•˜ì„¸ìš”.

ğŸ“Œ ì˜ˆì‹œ ì…ë ¥: `hello`
ğŸ“Œ ì˜ˆì‹œ ì¶œë ¥: `{'h':1, 'e':1, 'l':2, 'o':1}`

````python
text = input("ë¬¸ìì—´ì„ ì…ë ¥í•˜ì„¸ìš”: ")
count_char = {}

# textë¥¼ forë¡œ ìˆœíšŒ
for char in text:
    # ë¬¸ìê°€ ìˆìœ¼ë©´ dictì— +1
    if count_char.get(char, 0):
        count_char[char] += 1
    # ì—†ìœ¼ë©´ ìƒˆ í‚¤, count 1
    else:
        count_char.setdefault(char, 1)

print(count_char)
````

````python
# text = input("ë¬¸ìì—´ì„ ì…ë ¥í•˜ì„¸ìš”: ")
text = 'hello'
count_char = {}

# textë¥¼ forë¡œ ìˆœíšŒ
for char in text:
    # X in dict => keyì— ëŒ€í•´ í™•ì¸í•¨
    # ë¬¸ìê°€ ìˆìœ¼ë©´ ê·¸ í‚¤ì— +1
    if char in count_char:
        count_char[char] += 1
    # ì—†ìœ¼ë©´ ê°’ì„ 1ë¡œ ìƒˆë¡œ ë“±ë¡
    else:
        count_char[char] = 1

print(count_char)
````

````python
text = 'hello'
count_char = {}

for char in text:
    count_char[char] = count_char.get(char, 0) + 1

print(count_char)
````
